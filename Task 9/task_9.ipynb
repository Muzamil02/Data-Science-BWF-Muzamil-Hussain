# **Line Plot**

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 1, 2, 3, 4, 5]

plt.plot(x, y1, label='y = x^2', color='b', marker='o')
plt.plot(x, y2, label='y = x', color='r', linestyle='--')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

plt.legend()

plt.show()
```

# **Bar Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C']
values1 = [10, 20, 15]
values2 = [5, 15, 10]

x = np.arange(len(categories))
width = 0.35  # Width of the bars

fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, values1, width, label='Group 1')
bar2 = ax.bar(x + width/2, values2, width, label='Group 2')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Plot Example')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

# **Scatter Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.lmplot(x='total_bill', y='tip', data=tips, aspect=1.5)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot with Regression Line')

plt.show()
```

# **Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='YlGnBu')

plt.title('Heatmap of Monthly Flight Passengers')

plt.show()
```

# **Pair plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species')

plt.suptitle('Pair Plot Example', y=1.02)

plt.show()
```
